```python
from nltk.corpus import movie_reviews
```

```python
len(movie_reviews.fileids('pos'))
```

```python
movie_reviews.words(movie_reviews.fileids()[5])
```

```python
movie_reviews.words(movie_reviews.fileids()[1])
```

```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
```

```python
documents[0:5]
```

```python
import random
random.seed(0)
random.shuffle(documents)
```

```python
training_documents=documents[0:1500]
testing_documents=documents[1500:]
```

```python
import string
punc=list(string.punctuation)
stop=stop+punc
all_words=[]
for doc in training_documents:
    for w in doc[0]:
        if w.lower() not in stop:
            all_words.append(w.lower())
```

```python
len(all_words)
```

```python
import nltk
dist=nltk.FreqDist(all_words)
features=dist.most_common(3000)
```

```python
features
```

```python
features_word=[i[0] for i in features]
```

```python
features_word
```

```python
def get_features(document):
    words=set(document)
    features={}
    for w in features_word:
        features[w]=(w in words)
    return features    
    
```

```python
training_data=[(get_features(a),b) for (a,b) in training_documents]
testing_data=[(get_features(a),b) for (a,b) in testing_documents]
```

```python
classifier=nltk.NaiveBayesClassifier.train(training_data)
```

```python
classifier.show_most_informative_features(10)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```

```python
import pickle
```

```python
classifier_file=open("naivebayes.pickel","wb")
pickle.dump(classifier,classifier_file)
classifier_file.close()
```

```python
classifier_new=open("naivebayes.pickel","rb")
classifier_copy=pickle.load(classifier_new)
classifier_new.close()
```
